```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
import numpy
```

```python
common=Df1.keys().intersection(Df2.keys())
common_test=Df.keys().intersection(common)
```

```python
def nn_train(Df1,Df2,common,p,epoch,b_s):
    data=[]
    for j in common:
        
        ts1 =np.array(Df1[j])
        ts2=np.array(Df2[j])
        
       
        if sum(np.isnan(ts1))==0 and sum(np.isnan(ts2))==0:
            data+=[list(ts2[i:i+p])+list(ts1[i:i+p+1]) for i in range(len(ts1)-p)]
       
    np.random.shuffle(data)
    data=np.array(data)
    size=int(len(data)*0.8)
    train, test = data[0:size,:], data[size:len(data),:]
    X_train,X_test=train[:,0:-1],test[:,0:-1]
    y_train,y_test=train[:,-1],test[:,-1]
    model=Sequential()
   
    model.add(Dense(120, input_dim=2*p, activation='relu'))
    model.add(Dense(30, activation='relu'))
    model.add(Dense(1))
    # Compile model
    model.compile(loss='mean_squared_error',optimizer='adam')
    # Fit the model
    model.fit(X_train, y_train,validation_split=0.2, epochs=epoch, batch_size=b_s)
# evaluate the model
    scores = model.evaluate(X_test, y_test)
   
    return model
def nn_train2(Df1,Df2,common,p,epoch,b_s):
    data=[]
    for j in common:
        
        ts1 =np.array(Df1[j])
        ts2=np.array(Df2[j])
        
       
        if sum(np.isnan(ts1))==0 and sum(np.isnan(ts2))==0:
            data+=[list(ts2[i:i+p+1])+[ts1[i+p]] for i in range(len(ts1)-p)]
       
    np.random.shuffle(data)
    data=np.array(data)
    size=int(len(data)*0.8)
    train, test = data[0:size,:], data[size:len(data),:]
    X_train,X_test=train[:,0:-1],test[:,0:-1]
    y_train,y_test=train[:,-1],test[:,-1]
    model=Sequential()
   
    model.add(Dense(100, input_dim=p+1, activation='relu'))
    model.add(Dense(30, activation='relu'))
    model.add(Dense(1))
    # Compile model
    model.compile(loss='mean_squared_error',optimizer='adam')
    # Fit the model
    model.fit(X_train, y_train,validation_split=0.2, epochs=epoch, batch_size=b_s)
# evaluate the model
    scores = model.evaluate(X_test, y_test)
   
    return model
def nn_generalize(Df,Df1,nn,dataset,p):
    ts=Df[dataset]
    ts = np.array(ts)
    ts_pre=np.array(Df1[dataset])
   
    data=np.array([list(ts_pre[i:i+p])+list(ts[i:i+p+1]) for i in range(len(ts)-p)])
#np.random.shuffle(data)
    X_out,y_out=data[:,0:-1],data[:,-1]
    y_val=np.array(nn.predict(X_out)).reshape(1,-1)[0]
    #print(y_val,y_out)
    plt.figure(figsize=(15,10))
    plt.plot(y_out,'k')
    plt.plot(y_val,'r')
    plt.title(dataset+' RMSE='+str(np.sqrt(mean_squared_error(y_val,y_out))))
    plt.figure(figsize=(15,10))
    error=y_out-y_val
    plt.plot(error,'k')
    plt.title('Error Plot p='+str(p))
    print('Mean Percentage Error='+str(np.sqrt(mean_squared_error(y_out,y_val))/np.mean(y_out)))
def nn_generalize2(Df,Df1,nn,dataset,p):
    ts=Df[dataset]
    ts = np.array(ts)
    ts_pre=np.array(Df1[dataset])
   
    data=np.array([list(ts_pre[i:i+p+1])+[ts[i+p]] for i in range(len(ts)-p)])
#np.random.shuffle(data)
    X_out,y_out=data[:,0:-1],data[:,-1]
    y_val=np.array(nn.predict(X_out)).reshape(1,-1)[0]
    #print(y_val,y_out)
    plt.figure(figsize=(15,10))
    plt.plot(y_out,'k')
    plt.plot(y_val,'r')
    plt.title(dataset+' RMSE='+str(np.sqrt(mean_squared_error(y_val,y_out))))
    plt.figure(figsize=(15,10))
    error=y_out-y_val
    plt.plot(error,'k')
    plt.title('Error Plot p='+str(p))
    print('Mean Percentage Error='+str(np.sqrt(mean_squared_error(y_out,y_val))/np.mean(y_out)))
```
